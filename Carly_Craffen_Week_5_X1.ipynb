## Question 20

What fraction of the patients are `underweight`? What fraction of the patients are `overweight`? What fraction of the patients are `obese`? In order to answer these questions, you need to create a column for storing each patient's Body Mass Index (BMI). `BMI` is equal to $\dfrac{\text{weight in kilogram}}{\text{height in meters}^2}$. In calculating values in the new column of `BMI`, please make sure to use `.div` and `.pow` functions!

According to Center for Disease Control (CDC):
 - A `BMI` less than 18.5 falls within the `underweight` range.
 - A `BMI` between 25.0 and 30 (including 25 and excluding 30) is considered `overweight`.
 - A `BMI` of 30.0 or higher falls within the `obese` range.

**Hint** you need to use some of the below functions
 - `.eq()` for equality
 - `.ne()` for inequality
 - `.le()` for less than or equal 
 - `.lt()` for less than 
 - `.ge()` for greater than or equal 
 - `.gt()` for greater than 
 - `.between()` for between. between function has an amazing argument called `inclusive`. Read about it here: https://pandas.pydata.org/docs/reference/api/pandas.Series.between.html
 


```python
# solution
import pandas as pd
df_patients['BMI'] = df_patients.WEIGHT_KILOGRAM.div(df_patients.HEIGHT_METERS.pow(2))
underweight = df_patients.BMI.lt(18.5).mul(100).mean()
print(f'The fraction of the patients that are underweight is {underweight:.2f}%')
overweight = df_patients.BMI.between(25.0,30).mul(100).mean()
print(f'The fraction of the patients that are overweight is {overweight:.2f}%')
obese = df_patients.BMI.ge(30).mul(100).mean()
print(f'The fraction of the patients that are obese is {obese:.2f}%')
```

Then
 - In the first row, you need to plot a histogram (`hist` function), boxplot (`boxplot` function), and violinplot (`violinplot function)` of Statistical Analysis grades. 
 - In the second row, you need to plot a histogram (`hist` function), boxplot (`boxplot` function), and violinplot (`violinplot function)` of Python for Data Analysis grades.
